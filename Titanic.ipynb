```python
# pandas
import pandas as pd
from pandas import Series,DataFrame

# numpy, matplotlib
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

titanic_df = pd.read_csv("/home/rydrael/projects/ML-FirstSteps/train.csv", header = 0)
```

```python
titanic_df.head()
```

```python
# drop unnecessary columns, these columns won't be useful in analysis and prediction
titanic_df = titanic_df.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis = 1)
titanic_df.head()
```

Apply transformations to data so that it can be treated (Categorical features, etc)

```python
titanic_df[titanic_df["Sex"]=="male"]
```

```python
titanic_df["Age"] = titanic_df["Age"].fillna(titanic_df["Age"].median())
titanic_df.loc[titanic_df["Sex"]=="male","Sex"]=0
titanic_df.loc[titanic_df["Sex"]=="female","Sex"]=1
titanic_df.loc[titanic_df["Embarked"]=="S","Embarked"]=0
titanic_df.loc[titanic_df["Embarked"]=="C","Embarked"]=1
titanic_df.loc[titanic_df["Embarked"]=="Q","Embarked"]=2
titanic_df["Embarked"] = titanic_df["Embarked"].fillna(0)
```

```python
columns = list(titanic_df.columns[:9])
```

```python
from sklearn import tree

predictors = [col for col in columns if col != "Survived"]
X = titanic_df[predictors]
y = titanic_df["Survived"]

clf = tree.DecisionTreeClassifier()
clf.fit(X, y)

from IPython.display import Image  
from sklearn.externals.six import StringIO  
import pydot 

def printTree():
    dot_data = StringIO()  
    features = list(titanic_df.columns[:9])
    tree.export_graphviz(clf, out_file=dot_data,  
                             feature_names=features)  
    graph = pydot.graph_from_dot_data(dot_data.getvalue())  
    Image(graph.create_png())


printTree()
```
